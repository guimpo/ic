```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
np.set_printoptions(suppress=True)
pd.set_option('display.float_format', lambda x: '%.4f' % x)
```

```python
df_definition = pd.read_csv('data/exe_07_teste_rede_01.csv', sep=',', header=None)
definition = df_definition.iloc[0:20, [5]].values.astype(np.longdouble)

df_operation = pd.read_csv('data/exe_07_predicao_rede_01.csv', sep=',', header=None)
operation = df_operation.values.astype(np.longdouble)

fig, (ax0, ax1) = plt.subplots(1, 2, sharey='row')
fig.set_size_inches(8.5, 1.5)
fig.suptitle('Rede 01 - Treinamento 01', y=1.25)

ax0.xaxis.set_major_locator(MaxNLocator(integer=True))
ax0.set_title('Domínio de Definição')
ax0.plot([x for x in range(20)], definition, color='blue')

ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
ax1.set_title('Domínio de Operação')
ax1.plot([x for x in range(20)], operation, color='green', linestyle='--')

plt.show()
```
